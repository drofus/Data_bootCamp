## 🎲 Teoría de la probabilidad
La probabilidad nos ayuda a cuantificar la incertidumbre. En ciencia de datos, se utiliza para modelar fenómenos aleatorios y predecir resultados futuros. Entender la probabilidad es clave para tomar decisiones basadas en datos.

### 🔮 ¿Cómo se puede predecir el futuro?
Aunque no podemos predecir con certeza qué pasará, la **probabilidad** nos permite estimar qué tan probable es que ocurra un evento. Por ejemplo, podemos decir que hay un 70% de probabilidad de lluvia, lo que influye en nuestra decisión de llevar paraguas.

**Términos claves:**
|Término  |Definición |Ejemplo |
| :----------------: | :------: | :----: |
|Experimento        |Cualquier procedimiento que puede ser repetido.| Tirar un dado. 🎲|
|Prueba        |Una repetición de un experimento.   | Cada intento de tirar un dado. |
|Resultado     |Resultado de un experimento que se ejecuta una vez.  | Tirar un 1, 2, 3, 4, 5 o 6.|
|Evento        |Resultado único o conjunto de resultados a los que se puede asignar una probabilidad. | Tirar 5 o 6 (cualquier resultado posible que sea igual o mayor que 5). |

En estadística, utilizamos la letra P para denotar probabilidad y la letra A para denotar un evento de nuestro interés. Por lo tanto, la notación P(A) describe la probabilidad del evento A.
\begin{equation}
P(A) = \frac{n1}{n2}
\end{equation}
Siendo $n1$ el número de resultados que satisfacen el evento y $n2$ representa el número de resultados totales

### 🧩 Eventos independientes, multiplicación de probabilidades
Dos eventos son independientes si la ocurrencia de uno no afecta al otro. Por ejemplo, lanzar una moneda y lanzar un dado.

La probabilidad conjunta de eventos independientes se calcula multiplicando sus probabilidades individuales:

$$
P(A|B) = P(A)*P(B)
$$

```python
# Lanzar una moneda (P = 0.5) y un dado (P de 6 = 1/6)
prob_moneda = 0.5
prob_dado_6 = 1/6
prob_ambos = prob_moneda * prob_dado_6
print(f"Probabilidad de sacar cara y un 6: {prob_ambos:.3f}")
```

### 🧮 Fundamentos de combinatoria y otros problemas de probabilidad
La combinatoria nos permite contar posibles combinaciones sin tener que listarlas todas:
- **Permutaciones** (orden importa)
Una permutación es un arreglo de elementos en donde el orden importa. Para calcular el número de permutaciones de n elementos, los matemáticos utilizan la fórmula:
\begin{equation}
P_n = n!
\end{equation}
el signo "!" en expresiones matemáticas. Este signo se llama factorial y significa el producto de números naturales de 1 a n. En otras palabras, $n! = 1*2*3*...*(n-1)*n$
- **Combinaciones** (orden no importa)
son formas diferentes de agrupar elementos de un conjunto, donde la diferencia clave radica en si el orden de los elementos importa o no.  La fórmula para calcular el número de combinaciones es:
\begin{equation}
C_n^k = \frac{n!}{k!*(n-k)!}
\end{equation}

```python
import math
# ¿De cuántas formas se pueden elegir 2 estudiantes de un grupo de 5?
combinaciones = math.comb(5, 2)
print(f"Combinaciones posibles: {combinaciones}")
```

### 📊 La idea de distribución
Una distribución muestra cómo se distribuyen los valores de una variable aleatoria. Es fundamental para modelar fenómenos y hacer inferencias estadísticas.

Los tipos más comunes de distribuciones de probabilidad:
|Tipo de distribución |Descripción |Ejemplo |
| :----------------: | :------: | :----: |
|Distribución binomial  |La distribución de probabilidad discreta que describe el número de resultados positivos en un número fijo de ensayos.| Los resultados de la conversión en compra de llamadas publicitarias: Éxito: se realiza la compra. Fracaso: el cliente se niega a realizar la compra.|
|Distribución uniforme        |La distribución de probabilidad continua en la que cada valor dentro de un rango determinado tiene la misma probabilidad de ser seleccionado.   | Los resultados que se obtienen al tirar un dado: cada resultado tiene la misma probabilidad de ocurrir. |
|Distribución normal     |La distribución de probabilidad discreta o continua en la que los datos tienden a agruparse en torno a un valor medio o promedio.  | Resultados del examen SAT: la mayoría de los estudiantes de secundaria obtienen una puntuación media. Cuanto más cerca estén los resultados del mínimo o del máximo, con menos frecuencia se producirán.|

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(0, 10, size=1000)
plt.hist(data, bins=20, color='lightblue', edgecolor='black')
plt.title("Distribución Uniforme")
plt.grid()
plt.show()
```

```python
from scipy.stats import binom

n = 10  # número de pruebas
p = 0.5  # probabilidad de éxito

x = np.arange(0, 11)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color='orchid')
plt.title("Distribución Binomial: P(n=10, p=0.5)")
plt.xlabel("Número de caras")
plt.ylabel("Probabilidad")
plt.grid()
plt.show()
```

```python
from scipy.stats import norm

x = np.linspace(-4, 4, 100)
y = norm.pdf(x, loc=0, scale=1)

plt.plot(x, y, color='teal')
plt.title("Distribución Normal Estándar")
plt.grid()
plt.show()
```
