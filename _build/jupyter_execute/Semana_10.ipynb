En estad铆stica, utilizamos la letra P para denotar probabilidad y la letra A para denotar un evento de nuestro inter茅s. Por lo tanto, la notaci贸n P(A) describe la probabilidad del evento A.
\begin{equation}

\end{equation}
Siendo $n1$ el n煤mero de resultados que satisfacen el evento y $n2$ representa el n煤mero de resultados totales

### З Eventos independientes, multiplicaci贸n de probabilidades
Dos eventos son independientes si la ocurrencia de uno no afecta al otro. Por ejemplo, lanzar una moneda y lanzar un dado.

La probabilidad conjunta de eventos independientes se calcula multiplicando sus probabilidades individuales:

$$
P(A|B) = P(A)*P(B)
$$

```python
# Lanzar una moneda (P = 0.5) y un dado (P de 6 = 1/6)
prob_moneda = 0.5
prob_dado_6 = 1/6
prob_ambos = prob_moneda * prob_dado_6
print(f"Probabilidad de sacar cara y un 6: {prob_ambos:.3f}")
```

### М Fundamentos de combinatoria y otros problemas de probabilidad
La combinatoria nos permite contar posibles combinaciones sin tener que listarlas todas:
- **Permutaciones** (orden importa)
Una permutaci贸n es un arreglo de elementos en donde el orden importa. Para calcular el n煤mero de permutaciones de n elementos, los matem谩ticos utilizan la f贸rmula:
\begin{equation}
P_n = n!
\end{equation}
el signo "!" en expresiones matem谩ticas. Este signo se llama factorial y significa el producto de n煤meros naturales de 1 a n. En otras palabras, $n! = 1*2*3*...*(n-1)*n$
- **Combinaciones** (orden no importa)
son formas diferentes de agrupar elementos de un conjunto, donde la diferencia clave radica en si el orden de los elementos importa o no.  La f贸rmula para calcular el n煤mero de combinaciones es:
\begin{equation}
C_n^k = \frac{n!}{k!*(n-k)!}
\end{equation}

```python
import math
# 驴De cu谩ntas formas se pueden elegir 2 estudiantes de un grupo de 5?
combinaciones = math.comb(5, 2)
print(f"Combinaciones posibles: {combinaciones}")
```

Los tipos m谩s comunes de distribuciones de probabilidad:

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(0, 10, size=1000)
plt.hist(data, bins=20, color='lightblue', edgecolor='black')
plt.title("Distribuci贸n Uniforme")
plt.grid()
plt.show()
```

```python
from scipy.stats import binom

n = 10  # n煤mero de pruebas
p = 0.5  # probabilidad de 茅xito

x = np.arange(0, 11)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color='orchid')
plt.title("Distribuci贸n Binomial: P(n=10, p=0.5)")
plt.xlabel("N煤mero de caras")
plt.ylabel("Probabilidad")
plt.grid()
plt.show()
```

```python
from scipy.stats import norm

x = np.linspace(-4, 4, 100)
y = norm.pdf(x, loc=0, scale=1)

plt.plot(x, y, color='teal')
plt.title("Distribuci贸n Normal Est谩ndar")
plt.grid()
plt.show()
```
